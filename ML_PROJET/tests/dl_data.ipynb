```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Télécharger et préparer les données MNIST
(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.mnist.load_data()

# Normaliser les images
train_images = train_images.reshape(-1, 28*28) / 255.0
test_images = test_images.reshape(-1, 28*28) / 255.0

# Encodage one-hot des étiquettes
n_classes = 10
train_labels_onehot = np.zeros((train_labels.size, n_classes))
train_labels_onehot[np.arange(train_labels.size), train_labels] = 1
test_labels_onehot = np.zeros((test_labels.size, n_classes))
test_labels_onehot[np.arange(test_labels.size), test_labels] = 1
```

```python
# Afficher quelques exemples du jeu de données
plt.figure(figsize=(10, 2))
for i in range(10):
    plt.subplot(1, 10, i + 1)
    plt.imshow(train_images[i].reshape(28, 28), cmap='gray')
    plt.axis('off')
plt.show()
```
